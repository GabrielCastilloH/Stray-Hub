# DogFaceNet Training on Remote GPU

Quick training notebook for hackathon MVP.

**Time estimate**: 2-3 hours for 50 epochs on GPU

## Instructions:
1. Upload this notebook to JupyterHub
2. Upload `DogFaceNet_Dataset_224_1.zip` (72MB) to JupyterHub
3. Run all cells in order
4. Download `dogfacenet_weights.h5` when done
5. Place in `ml/checkpoints/` on your local machine
